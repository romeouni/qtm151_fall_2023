(c) Clean a column

<font size = "3">

- Our goal is to replace any string values in the column "position" <br>
to missing values (NaNs) and covert to numeric

- To do so
    1. Extract list of non-numeric values
    2. Replace certain list of values
    3. Convert column to numeric
    4. Display


```python
# Write your own code here
# Create and store a list of non-numeric values
non_numeric_values = sprint_results['position'][~sprint_results['position'].str.isnumeric()].tolist()
```

```python
# Create a list for replacement
sprint_results['position'].replace(non_numeric_values, float('nan'), inplace=True)
sprint_results['position'] = pd.to_numeric(sprint_results['position'], errors='coerce')

display(sprint_results['position'])
```

(d) Groupby + Aggregate

- Compute the mean and standard deviation of "position", <br>

```python
# Write your own code
df_aggContrcId_pos = sprint_results.groupby(by = 'constructorId')
df_aggConstrcId_pos = sprint_results.groupby('constructorId')['position'].agg(['mean', 'std'])

df_aggConstrcId_pos.sort_values(by=['mean'])

print(df_aggConstrcId_pos.iloc[0])
```
